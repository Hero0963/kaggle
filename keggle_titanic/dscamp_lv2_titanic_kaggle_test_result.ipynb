```python
import pandas as pd
import numpy as np
```

```python
df_test = pd.read_csv('https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/test.csv')
display(df_test)
```

```python
missing_values = df_test.isnull().sum()
display(missing_values)
```

```python
df_train = pd.read_csv('https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv')
```

```python
df_all = pd.concat([df_train, df_test], ignore_index=True)
```

```python
display(df_all)
```

```python
df_all['Title Row'] = df_all['Name'].map( lambda name: name.split( ',' )[1].split( '.' )[0].strip() )
```

```python
title_encoding = {
  "Capt":       "Officer",
  "Col":        "Officer",
  "Major":      "Officer",
  "Jonkheer":   "Royalty",
  "Don":        "Royalty",
  "Sir" :       "Royalty",
  "Dr":         "Officer",
  "Rev":        "Officer",
  "the Countess":"Royalty",
  "Dona":       "Royalty",
  "Mme":        "Mrs",
  "Mlle":       "Miss",
  "Ms":         "Mrs",
  "Mr" :        "Mr",
  "Mrs" :       "Mrs",
  "Miss" :      "Miss",
  "Master" :    "Master",
  "Lady" :      "Royalty"
}
```

```python
df_all['Title'] = df_all['Title Row'].map(title_encoding)
df_all['Title']
```

```python
df_all_Title_ohe = pd.get_dummies(df_all['Title'], prefix='Title_')
df_all_Title_ohe
```

```python
df_all = df_all.join(df_all_Title_ohe)
display(df_all)
```

```python
df_all['FamilySize'] = df_all['Parch'] + df_all['SibSp'] + 1
```

```python
df_all["Age"].fillna(df_all["Age"].median(), inplace=True)
df_all['Embarked'] = df_all['Embarked'].fillna(df_all['Embarked'].mode()[0])
df_all['Fare'] = df_all['Fare'].fillna(df_all['Fare'].median())
display(df_all.info())
```

```python
df_all_Embarked_ohe = pd.get_dummies(df_all['Embarked'], prefix='Embarked_')
df_all = df_all.join(df_all_Embarked_ohe)
```

```python
sex_encoding = {
    'male': 0,
    'female': 1
}
df_all['Sex'] = df_all['Sex'].map(lambda d: sex_encoding[d])
```

```python
display(df_all.columns)
```

```python
df_train_data_set = df_all[df_all['Survived'].notnull()]
df_test_data_set = df_all[df_all['Survived'].isnull()]

display(df_train_data_set.info())
display(df_test_data_set.info())
```

```python
columns_X = [
       'Pclass', 'Sex', 'Age', 'Fare',
       'SibSp', 'Parch',
       'Embarked__C', 'Embarked__Q', 'Embarked__S',
       'Title__Master', 'Title__Miss', 'Title__Mr', 'Title__Mrs', 'Title__Officer', 'Title__Royalty'
]
columns_y = ['Survived']


train_X = df_train_data_set[columns_X]
train_y = df_train_data_set[columns_y]

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

log = LogisticRegression(random_state=0, max_iter=3000)
scores = cross_val_score(log, train_X, train_y.values.ravel(),cv=5,scoring='accuracy')
print(scores)
print(scores.mean())
```

```python
# machine learning
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier , GradientBoostingClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100)
rf.fit(train_X, train_y.values.ravel())
```

```python
test_X = df_test_data_set[columns_X]
```

```python
predicted_labels = rf.predict(test_X)
submission = pd.DataFrame({'PassengerId': df_test_data_set['PassengerId'], 'Survived': predicted_labels})
submission['Survived'] = submission['Survived'].astype(int) # convert_data_type_important
submission.to_csv('dscamp_lv2_titanic_rf.csv', index=False)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.metrics import accuracy_score

class Model(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers, dropout_rate):
        super(Model, self).__init__()
        self.dropout_rate = dropout_rate
        self.num_layers = num_layers

        self.input_norm = nn.BatchNorm1d(input_size)

        self.hidden_layers = nn.ModuleList()
        self.hidden_layers.append(nn.Linear(input_size, hidden_size))

        for _ in range(num_layers - 1):
            self.hidden_layers.append(nn.Linear(hidden_size, hidden_size))

        self.predict = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.input_norm(x)

        for i in range(self.num_layers):
            x = self.hidden_layers[i](x)
            x = F.relu(x)
            x = F.dropout(x, p=self.dropout_rate, training=self.training)

        output = F.softmax(self.predict(x), dim=1)
        return output

# 模型、優化器初始化
input_size = train_X.shape[1]
hidden_size = 32
output_size = 2
num_layers = 2
dropout_rate = 0.1
model = Model(input_size, hidden_size, output_size, num_layers, dropout_rate)

optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.0) # 模型參數優化器
loss_func = torch.nn.CrossEntropyLoss()
```

```python
import tensorflow as tf
train_y_onehot = tf.keras.utils.to_categorical(train_y, num_classes=2)
```

```python
# 資料格式轉成torch專用格式
x_data = torch.tensor(train_X.values, dtype=torch.float32)
y_data = torch.tensor(train_y_onehot, dtype=torch.float32)
x_test_data = torch.tensor(test_X.values, dtype=torch.float32)
```

```python
batch_size = 16
num_epochs = 100
num_batches = len(train_X) // batch_size
loss_list = []
```

```python
for epoch in range(num_epochs):
    for i in range(num_batches):
        start = i * batch_size
        end = start + batch_size
        prediction = model(x_data[start:end])
        loss = loss_func(prediction, y_data[start:end])

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    loss = loss_func(model(x_data), y_data)
    loss_list.append(loss.item())




print(' ===== done =====')

print(loss_list)
```

```python
y_pred = model(x_test_data)
y_pred = y_pred.argmax(1) # one hot array to int array
```

```python
submission = pd.DataFrame({'PassengerId': df_test_data_set['PassengerId'], 'Survived': y_pred})
submission['Survived'] = submission['Survived'].astype(int) # convert_data_type_important
submission.to_csv('dscamp_lv2_titanic_dl.csv', index=False)
```
