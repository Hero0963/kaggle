```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv')
```

```python
display(df)
```

```python
missing_values = df.isnull().sum()
display(missing_values)
```

```python
data_types = df.dtypes
display(data_types)
```

```python
df_train = df.drop(['Name', 'Ticket', 'PassengerId', 'Cabin','SibSp', 'Parch'], axis=1)
display(df_train)
```

```python
# deal with missing values
df_train["Age"].fillna(df_train["Age"].median(), inplace=True)
df_train['Embarked'] = df_train['Embarked'].fillna(df_train['Embarked'].mode()[0])
display(df_train)
```

```python
# convert data type
s={'female':0,'male':1}
df_train['Sex'] = df_train['Sex'].map(s)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df_train['Embarked'] = label_encoder.fit_transform(df_train['Embarked'])
```

```python
display(df_train)
```

```python
bins = [0, 15, 55, float('inf')]
labels = ['Child', 'Adult', 'Elderly']
df_train['Age_category'] = pd.cut(df_train['Age'], bins=bins, labels=labels)
display(df_train)
```

```python
df_train['AgeLabel'] = label_encoder.fit_transform(df_train['Age_category'])
del df_train['Age_category']
del df_train['Age']
display(df_train)
```

```python
columns_X = set(df_train.columns) - {'Survived'}
columns_y = ['Survived']


train_X = df_train[columns_X]
train_y = df_train[columns_y]
```

```python
missing_values_X = train_X.isnull().sum()
display(missing_values_X)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

log = LogisticRegression(random_state=0, max_iter=3000)
scores = cross_val_score(log, train_X, train_y.values.ravel(), cv=5, scoring='accuracy')
print(scores)
```
